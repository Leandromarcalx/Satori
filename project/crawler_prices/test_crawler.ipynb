```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def crawl_website(root_url, max_pages=100):
    visited = set()
    to_visit = [root_url]
    i = 0
    while to_visit and len(visited) < max_pages:
        print('still working', i)
        i += 1
        url = to_visit.pop(0)
        if url in visited:
            continue

        try:
            response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
            if response.status_code != 200:
                continue

            visited.add(url)
            soup = BeautifulSoup(response.text, 'html.parser')

            for link in soup.find_all('a', href=True):
                href = link['href']
                full_url = urljoin(root_url, href)
                parsed_root = urlparse(root_url).netloc
                parsed_link = urlparse(full_url).netloc

                # Stay within the same domain
                if parsed_root == parsed_link and full_url not in visited:
                    to_visit.append(full_url)

        except Exception as e:
            print(f"Failed to access {url}: {e}")

    return list(visited)

# Example usage
root_website = "https://www.zara.com/br/"
pages = crawl_website(root_website)
for page in pages:
    print(page)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_prices_and_categories(url):
    # Fetch website content

    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    }

    response = requests.get(url, headers=headers)
    
    soup = BeautifulSoup(response.text, 'html.parser')

    # Replace with actual class or tag names from the website
    price_elements = soup.find_all(class_ = 'product-grid-product-info__product-price price')
    product_elements = soup.find_all(class_ = 'product-grid-product-info__main-info')

    prices = [price.get_text(strip=True) for price in price_elements]
    categories = [category.get_text(strip=True) for category in product_elements]

    # Combine prices and categories
    data = list(zip(categories, prices))

    return data


# url = 'https://www.zara.com/br/pt/woman-sale-l5503.html?v1=2364041'
lst = []

for url in pages:
    result = scrape_prices_and_categories(url)

    lst.append(result)

print(len(lst))
```

```python
df = pd.DataFrame(lst)
```
