Class Number 2 - Behind the Pipeline

```python
from transformers import pipeline
from transformers import AutoTokenizer
```

```python
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    "Understading a little bit more about AI makes me motivated.", 
    "Not knowing about how hidden layers work frustrates me."
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
from transformers import AutoModel

model = AutoModel.from_pretrained(checkpoint)
```

```python
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
```

```python
print(outputs.logits.shape)
```

```python
print(outputs.logits)
```

```python
import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

Tokenziers    

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = "Using a Transformer network is simple"
tokens = tokenizer.tokenize(sequence)

print(tokens)
```

```python
ids = tokenizer.convert_tokens_to_ids(tokens)

print(ids)
```

```python
final_inputs = tokenizer.prepare_for_model(ids)
print(final_inputs['input_ids'])
```

```python
decoded_string = tokenizer.decode(final_inputs['input_ids'])
print(decoded_string)
```
