```python
def get_high_spenders(transactions):
    user_spend = {}

    # 1. Calculate total spent per user
    for user_id, _, price in transactions:
        if user_id in user_spend:
            user_spend[user_id] += price
        else:
            user_spend[user_id] = price

    # 2. Calculate average spending
    totals = list(user_spend.values())
    avg = sum(totals) / len(totals)

    # 3. Filter users who spent above average
    result = [user for user in user_spend if user_spend[user] > avg]

    return sorted(result)
```

```python
get_high_spenders(transactions)
```
